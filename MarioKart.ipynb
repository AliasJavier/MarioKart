# Importamos librerias

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#data = pd.read_csv("KartKross.csv")
data = pd.read_csv("https://raw.githubusercontent.com/AliasJavier/MarioKart/main/KartKross.csv")
df = pd.DataFrame(data, columns=['characters','kart','tires','glider','speed','acceleration','weight','handling','traction','turbo','total']
)
data.dtypes
```

```python
fig, axs = plt.subplots(3, figsize = (5,5))
plt1 = sns.boxplot(data['weight'], ax = axs[0])
plt2 = sns.boxplot(data['speed'], ax = axs[1])
plt3 = sns.boxplot(data['acceleration'], ax = axs[2])
plt.tight_layout()
```

```python
sns.pairplot(data, x_vars=['weight', 'speed', 'acceleration'], y_vars='handling', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
sns.heatmap(data.corr(), cmap="Greens", annot = True)
plt.show()
```

```python
sns.pairplot(data)
```

```python
corrmat = data.corr()
hm = sns.heatmap(corrmat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=df.columns, 
                 xticklabels=df.columns, 
                 cmap="Spectral_r")
plt.show()
```
